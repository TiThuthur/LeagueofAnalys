```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.core.interchange.dataframe_protocol import DataFrame
import numpy as np
bans : DataFrame = pd.read_csv("data/bans.csv")
rumble:int = bans.apply(lambda col: col.str.contains('Rumble')).sum().sum()
kassadin:int = bans.apply(lambda col: col.str.contains('Kassadin')).sum().sum()

display(rumble)
rumble_array  = np.array(rumble)
limit  = np.array(2000)
plt.plot(rumble_array,limit, color='red')

plt.title("Nombre de bans pour Rumble")
plt.xlabel("Rumble")
plt.ylabel("nombre")

plt.show()
```
